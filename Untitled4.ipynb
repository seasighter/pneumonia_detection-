```python
# loading data with tf.keras api
from tensorflow.keras.preprocessing import image_dataset_from_directory

# to load the dataset directory in x,y x will be iamge and y wil be label
train_ds=image_dataset_from_directory(
    'chest_xray/train',
    labels='inferred',
    label_mode='int',
    class_names=['NORMAL','PNEUMONIA'],  # [0 is normal and 1 is pneumonia]
    color_mode='grayscale',
    batch_size=32,
    image_size=(64,64),
    shuffle=True,
    seed=200, # how much to shuffle togethher
    validation_split=None, # not seperating as we have validation data seperated already
    subset=None,
    )
# train_ds=(batch_size,size_of_img,size_of_img,no_of channel)
test_ds = image_dataset_from_directory(
    'chest_xray/test',
    labels='inferred',
    label_mode='int',
    class_names=['NORMAL','PNEUMONIA'],
    color_mode='grayscale',
    batch_size=32,
    image_size=(64,64),
    shuffle=True,
    seed=100,
    validation_split=None,
    subset=None,
)

val_ds = image_dataset_from_directory(
    'chest_xray/val',
    labels='inferred',
    label_mode='int',
    class_names=['NORMAL','PNEUMONIA'],
    color_mode='grayscale',
    batch_size=32,
    image_size=(64,64),
    shuffle=True,
    seed=4,
    validation_split=None,
    subset=None,
    )
```

```python
val_ds
```

```python
# visualizing the data
plt.figure(figsize=(10,10))
for images ,labels in train_ds.take(1):
  for i in range(9):
    ax=plt.subplot(3,3,i+1)
    plt.imshow(images[i].numpy().astype("uint8"),cmap="gray")
    plt.title(int(labels[i]))
    plt.axis("off")
```

```python
from tensorflow.keras.layers import Rescaling
i=Input(shape=(64,64,1))
x=Rescaling(scale=1.0/255)(i) # to rescale the data
x=Conv2D(filters=32,kernel_size=(3,3),strides=(1,1),padding='same',activation='relu')(x)
x=Conv2D(filters=32,kernel_size=(3,3),strides=(1,1),padding='same',activation='relu')(x)  # reduced the architecture becuz it was too big for free version
x=Dropout(0.2)(x)
x=Conv2D(filters=64,kernel_size=(3,3),strides=(1,1),padding='same',activation='relu')(x)
x=Dropout(0.2)(x)
x=Conv2D(filters=128,kernel_size=(3,3),strides=(1,1),padding='same',activation='relu')(x)
x=Flatten()(x)
x=Dense(224,activation='relu')(x)
x=Flatten()(x)
x=Dropout(0.2)(x)
x=Dense(2,activation='softmax')(x)
model=Model(i,x)
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
r=model.fit(train_ds,epochs=15,validation_data=val_ds,shuffle=True)
```

```python
# plot the curve for training and validation loss
plt.plot(r.history["loss"],label="loss")
plt.plot(r.history["val_loss"],label="validation_loss")
plt.legend();
```
